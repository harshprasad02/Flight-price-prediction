```python
driver = webdriver.Chrome("chromedriver.exe")
```

```python
url = "https://flight.yatra.com/air-search-ui/dom2/trigger?ADT=1&CHD=0&INF=0&class=Economy&destination=BOM&destinationCountry=IN&flexi=0&flight_depart_date=05%2F10%2F2021&hb=0&noOfSegments=1&origin=DEL&originCountry=IN&type=O&unique=514688326983&viewName=normal"
driver.get(url)
```

```python
name = []          #name of flight        
date = []          #date of flight         
source = []        #source of the flight        
destination = []   #destination of the flight                   
dep_time = []      #departure time of the flight           
arr_time = []     #arrival time of the flight            
duration = []     #duration of the flight      
stops = []        #number of stops
price = []      #price od flight ticket
```

```python
#6th october
#name
try:
    all_names = driver.find_elements_by_xpath("//div[@class = 'fs-13 airline-name no-pad col-8']//span[@class = 'i-b text ellipsis']")

    for i in all_names:
        name.append(i.text)

except NoSuchElementException:
    name.append('NaN')
    
#date
try:
    all_dates = driver.find_element_by_xpath("//ul[@class = 'full-width no-wrap ovf-hidden']//li[2]//div//p[1]")
    linked =  driver.find_elements_by_xpath("//div[@class = 'fs-13 airline-name no-pad col-8']//span[@class = 'i-b text ellipsis']")
    for j in range(0,len(linked)):
        date.append(all_dates.text)
except NoSuchElementException:
    date.append('NaN')
    
    
#dep_time
try:
    dep_times = driver.find_elements_by_xpath("//div[@class = 'i-b col-4 no-wrap text-right dtime col-3']//div//div[@class = 'i-b pr']")

    for i in dep_times:
        dep_time.append(i.text)
except NoSuchElementException:
    dep_time.append('NaN')

    
#arr_time
try:
    arr_times = driver.find_elements_by_xpath("//div[@class = 'i-b pdd-0 text-left atime col-5']//p[1]")

    for i in arr_times:
        arr_time.append(i.text)
except NoSuchElementException:
    arr_time.append('NaN')
    

#source
try:
    sources = driver.find_elements_by_xpath("//div[@class = 'i-b col-4 no-wrap text-right dtime col-3']//p[@class = 'fs-10 font-lightgrey no-wrap city ellipsis']")

    for i in sources:
        source.append(i.text)
except NoSuchElementException:
    source.append('NaN')
    
    
#destination
try:
    destinations = driver.find_elements_by_xpath("//div[@class = 'i-b pdd-0 text-left atime col-5']//p[@class = 'fs-10 font-lightgrey no-wrap city ellipsis']")

    for i in destinations:
        destination.append(i.text)
except NoSuchElementException:
    destination.append('NaN')
    
#duration
try:
    durations = driver.find_elements_by_xpath("//div[@class = 'stop-cont pl-13']//p[@class = 'fs-12 bold du mb-2']")

    for i in durations:
        duration.append(i.text)
except NoSuchElementException:
    duration.append('NaN')  
    
    
#stops
try:
    stopss = driver.find_elements_by_xpath("//div[@class = 'stop-cont pl-13']//div[@class = ' font-lightgrey fs-10 tipsy i-b fs-10']//span")

    for i in stopss:
        stops.append(i.text)
except NoSuchElementException:
    stops.append('NaN')
    
    
#price
try:
    fares = driver.find_elements_by_xpath("//div[@class = 'i-b tipsy fare-summary-tooltip fs-18']")

    for i in fares:
        price.append(i.text)
except NoSuchElementException:
    price.append('NaN')
```

## makemytrip

```python
driver2 = webdriver.Chrome("chromedriver.exe")
```

```python
url = "https://www.makemytrip.com/flights/"
driver2.get(url)
```

```python
name3 = []          #name of flight        
date3 = []          #date of flight         
source3 = []        #source of the flight        
destination3 = []   #destination of the flight                   
dep_time3 = []      #departure time of the flight           
arr_time3 = []     #arrival time of the flight            
duration3 = []     #duration of the flight      
stops3 = []        #number of stops
price3 = []      #price od flight ticket
```

```python
#makemytrip we run the same code for different days and different cities

#names
try:
    all_names = driver2.find_elements_by_xpath("//div[@class = 'makeFlex align-items-center  ']//span")

    for i in all_names:
        name3.append(i.text)

except NoSuchElementException:
    name3.append('NaN')
    
#date
try:
    all_dates = driver2.find_element_by_xpath("//div[@class = 'slider-list']//div[69]//a//p[1]")
    linked =  driver2.find_elements_by_xpath("//div[@class = 'makeFlex align-items-center  ']//span")
    for i in range(0,len(linked)):
        date3.append(all_dates.text)
except NoSuchElementException:
    date3.append('NaN')

    
    
#dep_time
try:
    dep_times = driver2.find_elements_by_xpath("//div[@class = 'flightTimeSection flexOne timeInfoLeft']//p//span")
    for i in dep_times:
        dep_time3.append(i.text)
except NoSuchElementException:
    dep_time3.append('NaN')

    
#arr_time
try:
    arr_times = driver2.find_elements_by_xpath("//div[@class = 'flightTimeSection flexOne timeInfoRight']//p//span")

    for i in arr_times:
        arr_time3.append(i.text)
except NoSuchElementException:
    arr_time3.append('NaN')
    

#source
try:
    sources = driver2.find_elements_by_xpath("//div[@class = 'flightTimeSection flexOne timeInfoLeft']//p[@class = 'darkText']")
    for i in sources:
        source3.append(i.text)
except NoSuchElementException:
    source3.append('NaN')
    
    
#destination
try:
    destinations = driver2.find_elements_by_xpath("//div[@class = 'flightTimeSection flexOne timeInfoRight']//p[@class = 'darkText']")
    
    for i in destinations:
        destination3.append(i.text)
except NoSuchElementException:
    destination3.append('NaN')
    
#duration
try:
    durations = driver2.find_elements_by_xpath("//div[@class = 'stop-info flexOne']//p")

    for i in durations:
        duration3.append(i.text)
    
except NoSuchElementException:
    duration3.append('NaN')
    
    
#stops
try:
    stopss = driver2.find_elements_by_xpath("//div[@class = 'stop-info flexOne']//div//p[@class = 'flightsLayoverInfo']")

    for i in stopss:
        stops3.append(i.text)
except NoSuchElementException:
    stops3.append('NaN')
    
    
#price
try:
    fares = driver2.find_elements_by_xpath("//div[@class = 'textRight flexOne']//p")

    for i in fares:
        price3.append(i.text)
except NoSuchElementException:
    price3.append('NaN')
```

```python
#combining all the lists together

#name

name.extend(name2)
name.extend(name3)

#date
date.extend(date2)
date.extend(date3)
#source
source.extend(source2)
source.extend(source3)
#destnation
destination.extend(destination2)
destination.extend(destination3)
#dep_time
dep_time.extend(dep_time2)
dep_time.extend(dep_time3)
#arr_time
arr_time.extend(arr_time2)
arr_time.extend(arr_time3)
#duration
duration.extend(duration2)
duration.extend(duration3)
#stops
stops.extend(stops2)
stops.extend(stops3)
#price
price.extend(price2)
price.extend(price3)
```

```python
Flight_price = pd.DataFrame({"Airline Name" : name , "Date of journey" : date , "Source" : source , "Destination": destination , 
                          "Departure Time" : dep_time , "Arrival Time" : arr_time , "Duration" : duration ,
                         "Total Stops" : stops , "Price" : price})
```
